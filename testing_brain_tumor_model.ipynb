```python
from tensorflow.keras.models import load_model
```

```python
model = load_model('brain_tumor_detect_model_trained.h5')
```

```python
import os
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
```

```python
os.listdir(data_path + '/Testing')
```

```python
glioma_cell = imread(data_path + '/Testing/' + '/glioma_tumor/' + 'image(1).jpg')
```

```python
plt.imshow(glioma_cell)
```

```python
import cv2
```

```python
glioma_cell = cv2.resize(glioma_cell, (400, 400))
```

```python
glioma_cell.shape
```

```python
glioma_cell = np.expand_dims(glioma_cell, axis=0)
```

```python
r = np.argmax(model.predict_generator(glioma_cell))
```

```python
r
```

```python
# {'glioma_tumor': 0, 'meningioma_tumor': 1, 'no_tumor': 2, 'pituitary_tumor': 3}

for i in os.listdir(data_path + '/Testing/' + '/pituitary_tumor'):
    
    img = imread(data_path + '/Testing/' + '/pituitary_tumor/' + i)
    
    img = cv2.resize(img, (400, 400))
    
    img = np.expand_dims(img, axis=0)
    
    res = np.argmax(model.predict_generator(img))
    
    print(res)
```
